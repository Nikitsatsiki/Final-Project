# Extract Metadata of an Image
#### A lot of images, especially when downloaded from the web are already processed and contain no metadata, so the dictionary would be just `None`-Type and would contain no information. <br/> I personally managed to get metadata of images that I took by myself with my own camera.

```python
#import numpy as np
#import pandas as pd
#import os
from PIL import Image, ExifTags
#import pywt
#import matplotlib.pyplot as plt
```

```python
#import image
sea = Image.open('DSC00309.jpg')
#get metadata
exif_data = sea._getexif()
#create a usable dictionary
exif = {ExifTags.TAGS[k]: v for k, v in sea._getexif().items() if k in ExifTags.TAGS}
```

```python
print(type(exif_data), type(exif))
```
