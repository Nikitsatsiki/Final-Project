```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
sea = Image.open("data/resolution/DSC05820.jpg");
```

```python
sea.size #scale new image in the same order (e.g. 4.8 * 2.7)
         #the size must be integer, so I go with 5 * 3 and multiples
```

```python
fig,ax=plt.subplots(figsize=(5, 3))
ax.imshow(sea);
```

```python
sea_reduced = sea.resize((15, 9), Image.ANTIALIAS) #ANTIALIAS: a high-quality downsampling filter
np.asarray(sea_reduced)
```

```python
fig, ax = plt.subplots(figsize=(30, 50))
plt.imshow(sea_reduced);
```

```python
sea_reduced.save('..\presentation\sea_reduced.jpg')
```
